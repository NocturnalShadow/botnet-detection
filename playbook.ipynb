```python
import pandas as pd
import numpy as np

# Load flows
flows = pd.read_csv('ISCX_Botnet-Training.pcap_Flow.csv', index_col=False)
# flows = pd.read_csv('ISCX_Botnet-Testing.pcap_Flow.csv', index_col=False)

# Filter flows
flows = flows[flows["Protocol"].isin(["TCP", "UDP", "ICMP"])]
flows = flows.replace(np.nan, 0).replace(np.inf, 0)
flows = flows[flows["Flow Duration"] >= 0]

flows
```

```python
# Filter features
non_feature = ["Flow ID", "Src IP", "Src Port", "Dst IP", "Dst Port", "Timestamp", "HostType", "FlowType", "Protocol", "FlowTerminationReason"]
flows = flows[flows.columns.drop(list(flows.filter(regex='Idle')))]
flows_features = flows[flows.columns.drop(non_feature)]
flow_labels = flows["FlowType"].replace(to_replace=r"^(.(?<!BENIGN))*?$", value="MALICIOUS",regex=True).replace({ "BENIGN": 0, "MALICIOUS": 1 })
flows_features
```

```python
from pickle import load

# Load scaler
scaler = load(open("scaler.pkl", 'rb'))
```

```python
import tensorflow as tf

# Load model
dnn = tf.keras.models.load_model("dnn_2epochs.h5")
dnn.summary()
```

```python
# Scale features
scaled_features = scaler.transform(flows_features.to_numpy())

# Predict
predictions = dnn.predict(scaled_features)
predictions
```

```python
from sklearn.metrics import roc_auc_score, accuracy_score

# Estimate performance
treshold = 0.5
print("DNN (16x16x16):")
print(" AUROC: " + str(roc_auc_score(flow_labels, predictions)))
print(" Accuracy: " + str(accuracy_score(flow_labels, [probability > treshold for probability in predictions])))
```
